```python
import requests

def get_weather(city):
    api_key = '50a43125580e418f981b0f4572be0c78'  # Replace with your actual API key
    url = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric'

    response = requests.get(url)
    if response.status_code == 200:
        return response.json()  # Return the API response as a JSON object
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return None
```

```python
# Test the function with an example city
city_weather = get_weather('hubli')

if city_weather:
    print(city_weather)  # Print the entire response to inspect it
```

```python
if city_weather:
    city = city_weather.get('name', 'Unknown City')
    weather = city_weather['weather'][0]['description'] if 'weather' in city_weather else 'No Weather Data'
    temperature = city_weather['main']['temp'] if 'main' in city_weather else 'No Temperature Data'

    print(f"City: {city}")
    print(f"Weather: {weather}")
    print(f"Temperature: {temperature} °C")
```
